# Import stuff
Import libraries used to make the plots

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

# Load and analyse pos_full
Read train_pos_full.txt and count the number of words per tweet.

```
with open('twitter-datasets/train_pos_full.txt') as f:
  content = f.readlines()
content = [x.strip() for x in content]
```

```
stat = []
for line in content:
  counter = 0
  for letter in line:
    if letter == ' ':
      counter += 1
  counter +=1
  stat.append(counter)
pos_stat = pd.DataFrame(stat)
```

# Plot pos_full
Make a histogramm of the length of the tweets in train_pos_full.txt and also print the mean length, the standard deviation, the minimum length and the maximum length.

```
pos_mean = pd.DataFrame.mean(pos_stat)[0]
pos_min = pd.DataFrame.min(pos_stat)[0]
pos_max = pd.DataFrame.max(pos_stat)[0]
pos_std = pd.DataFrame.std(pos_stat)[0]
print("Mean tweet length [words]: " + str(pos_mean))
print("Std: " + str(pos_std))
print("Min: " + str(pos_min))
print("Max: " + str(pos_max))
sns.set(color_codes=True)
sns.distplot(pos_stat, kde=False)
plt.xlim(pos_min, pos_max)
```

# Combine plots
Combine the histogramms of the positive and negative tweets for both the original data as well as the extra data.

```
sns.set(color_codes=True)
plot=sns.distplot(pos_extra_stat, label='Extra Data Pos', kde=False)
plot=sns.distplot(neg_extra_stat, label='Extra Data Neg', kde=False)
plt.legend()
plot.set(xlabel='length [Words]')
plt.xlim(0, 40)
plt.show()
fig = plot.get_figure()
fig.savefig('Extra_data.png')

plot=sns.distplot(pos_stat, label='Pos Full', kde=False)
plot=sns.distplot(neg_stat, label='Neg Full', kde=False)
plt.legend()
plot.set(xlabel='length [Words]')
plt.xlim(0, 40)
plt.show()
fig = plot.get_figure()
fig.savefig('Base_data.png')
```
